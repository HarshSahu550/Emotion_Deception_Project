```python
import pandas as pd
```

```python
df = pd.read_csv("age_gender.csv")
df.head()
```

```python
df['pixels'][0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

pixels = df.loc[40, 'pixels']

img = np.fromstring(pixels, dtype=np.uint8, sep=' ').reshape(48, 48)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import cv2
import os

for i in range(len(df['pixels'])):
    pixels = df.loc[i, 'pixels']

    img = np.fromstring(pixels, dtype=np.uint8, sep=' ').reshape(48, 48)

    # create folder if not exists
    os.makedirs("images", exist_ok=True)

    # save image
    cv2.imwrite(f"images/img{i}.png", img)
```
